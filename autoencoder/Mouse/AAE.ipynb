```python
import autoencoder.adversarial as aae

from matplotlib import pyplot as plt
%matplotlib inline
```

```python
def setup():
    esize = [(128,3), (256, 3), (512,3)]
    dsize = list(reversed(esize))

    params =dict()
    
    params['width'] = 32
    params['height'] = 32
    params['nchannels'] = 4
    params['channels'] = [0,1,3,4]
    params['nepochs'] = 20
    params['batchsize'] = 256
    params['learning_rate'] = 0.0003
    params['restore'] = False
    params['latent_size'] = 256
    params['enc_sizes'] = esize 
    params['dec_sizes'] = dsize
    params['droprate'] = 0.85
    params['stdev'] = 0.04
    params['denoise'] = False
    params['slam'] = 0

    return params
```

```python
datadir = "/media/cjw/Data/cyto/Mouse/mmCompensated/"
aae.train(132001, params=setup(), datadir=datadir, display=True, display_int=1000, report_int=100,
          title="Adver_mouse_256")
```
